### Gadfly ABC
- the type of plot is decided by `Geom` (Geometry) - by default `Geom.point`
- the needed data `aesthetics` is decided by `Geom`
- decorations are specified by `scales`, `coordinates`, and `guides`
- a single plot can have multiple `layers`
- the plotting in Gadfly is mainly done by invoking `plot`
- other types of plotting
    - plotting a dataframe: dataframe is the first param, aesthetics should be columns
    - plotting functions/expressions: `plot(functions, a, b)` or `@plot(expr, a, b)`
    - plotting to multiple layers: use `layer()` to wrap layers
- use `draw` to dump the plot to other formats such as png, pdf and d3js

```julia
using Gadfly
```

```julia
## default Geom.point needs both x and y for aesthetics
plot(x = rand(10), y = rand(10))
```

```julia
## now Gemo.point and Gemo.line, both uses x and y
plot(x = rand(10), y = rand(10), Geom.point, Geom.line)
```

```julia
## decorate the previous plotting 
fig = plot(x = 1:10, y = 2.^rand(10), Scale.y_log, Geom.point, Geom.smooth, 
Guide.xlabel("Input"), Guide.ylabel("Output"), Guide.title("Dog Training"))
```

## output to different plotting format

```julia
draw(PNG("myplot.png", 4inch, 3inch), fig)
```

```julia
draw(SVGJS("myplot.js.svg", 4inch, 3inch), fig)
```

```julia
## plot DataFrames
## color, like x, y, is another type of aesthetic
using RDatasets
iris = dataset("datasets", "iris")
plot(iris, x="SepalLength", y="SepalWidth", color = "Species")
```

```julia
@manipulate for x in names(iris), y in names(iris)
    plot(iris, x=x, y=y, color = "Species")
end
```

```julia
plot(iris, x = "SepalLength", color = "Species", Geom.histogram)
```

```julia
## functions and expressions

plot([sin, cos], -1, 1)
```

```julia
@plot(2*x/cos(x), -1, 1)
```

```julia
plot(x -> 2*x/cos(x), -1, 1)
```

```julia
## customized legends
plot([sin, cos], -1, 1, color=["sin", "cos"])
```

```julia
## customized label and colors
plot([sin, cos], -1, 1, color=["sin", "cos"], 
Scale.discrete_color_manual("blue", "green"))
```

```julia
## multi layers
plot(layer(x = rand(10), y = rand(10), Geom.point, 
        Theme(default_color = color("blue"))), 
    layer(x = rand(10), y = rand(10), Geom.point, 
        Geom.line, Geom.smooth, Theme(default_color = color("red"))))
```

```julia
## multiplayer for same dataframe
plot(iris, layer(x = "SepalLength", y="SepalWidth", Geom.point), 
layer(x="PetalLength", y = "PetalWidth", Geom.line))
```

```julia
## subplots
hstack(
    plot(iris, y = "SepalLength", Geom.boxplot),
    plot(iris, y = "SepalWidth", Geom.boxplot),
    plot(iris, x = "Species", y = "PetalLength", Geom.boxplot)
)
```
